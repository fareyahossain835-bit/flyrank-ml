## 3. Quick look at the data (2-3 real numbers)

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Real numbers from the dataset
corr = df["search_volume"].corr(df["impressions_90d"])
rows = len(df)
columns = len(df.columns)
median_down = df[df["trend_direction"]=="down"]["word_count"].median()

print("Rows:", rows)
print("Columns:", columns)
print("Correlation (search_volume vs impressions_90d):", round(corr,6))
print("Median word count for declining pages:", median_down)



from IPython.display import Markdown, display

display(Markdown("""
**Observations:**

**The starter dataset contains 30,000 rows.**

**The correlation between search volume and impressions is very low (≈0.0012).**

**The median word count of declining pages is about 2909 words.**

**These observations suggest that simple metrics alone may not explain page performance, making this lane suitable for further analysis.**



"""))
```
